```python
def save_plot(x_input,n=4):
    for i in range(n*n):
        plt.subplot(n, n, i+1)
        plt.imshow(x_input[i,:,:,:])
        plt.axis('off')
    plt.show()
```

```python
def animo_image_generation(model):
    latent_dim = 512
    n_samples = 25
    x_input = randn(latent_dim * n_samples)
    z_input = x_input.reshape(n_samples, latent_dim)
    X  = model.predict(z_input)
    # scale from [-1,1] to [0,1]
    X = (X + 1) / 2.0
    save_plot(X, n=5)
```

```python
generator = tf.keras.models.load_model("generator_model1242.h5")
animo_image_generation(generator)
```
